```python
from google import genai
from google.genai import types
from dotenv import load_dotenv
import os
import pandas as pd
import json
import time
```

```python
df = pd.read_csv("combined_user_comments.csv")

df.head()
```

```python
load_dotenv()
gemini_api = os.getenv("Gemini_API") 

# The client gets the API key from the environment variable `GEMINI_API_KEY`.
client = genai.Client(api_key=gemini_api)
```

```python
count = int(df["comment"].count())
count =  count/3
```

```python
prompt = """
        You are an AI system evaluating Instagram comments for influencer audience quality analysis.

        Your task is to classify a comment into ONE of the following categories:

        1. RELEVANT
            1. Purchase Intent
                - Clear intent to buy or ask how to buy

            2. Product Inquiry
                - Questions about price, availability, outlets, sizing

            3. Product Feedback
                - Opinions, reviews, or experiences with the product

            4. Brand-Relevant Discussion
                - Mentions brand, collection, store, campaign, or product line

        2. IRRELEVANT
            1. Noise / Low Quality
                - Emojis only, generic praise, spam, self-promotion, irrelevant content


        Return ONLY:
        RELEVANT OR IRRELEVANT
        """


def analyze_comment(comment):        
    response = client.models.generate_content(
        model="gemini-3-flash-preview",
        config=types.GenerateContentConfig(
        system_instruction=prompt),
        contents=comment
    )
    return (response.text)
```

```python
delay = time.sleep(10)
count_relevant_comment = 0

for i in range(0, count):
    comment =df["comment"][i]
    output = analyze_comment(comment)
    if(output == "RELEVANT"):
        count_relevant_comment+=1
    delay

print(count_relevant_comment)
```

```python
Auidence_quality = (count_relevant_comment/count)*100
print(Auidence_quality)
```
