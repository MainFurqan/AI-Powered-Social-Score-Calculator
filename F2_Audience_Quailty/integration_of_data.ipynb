```python
import pandas as pd 
```

```python
df_1 = pd.read_csv('dataset_instagram-scraper.csv')
df_1.head(10)
```

```python
print(df_1.shape)
df_1.columns
```

```python
df_2 = pd.read_csv('my_extracted_comments.csv')
print(df_2.shape)
df_2.columns
```

```python
# Extract relevant columns from each
df1_subset = df_1[['owner/username', 'text']].copy()
df1_subset.columns = ['username', 'comment']  # Rename to match df_2

df2_subset = df_2[['username', 'comment']].copy()

# Concatenate rows from both DataFrames
combined_df = pd.concat([df2_subset, df1_subset], ignore_index=True)

# Save to new CSV
combined_df.to_csv('combined_user_comments.csv', index=False)
```
