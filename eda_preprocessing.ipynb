```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("mahirahkhan_instagram_data.csv")
df.head()
```

```python
df.count()
```

```python
views = ['1.5M', '2.2M', '471K', '1.5M', '909K', '378K', '406K', '500K', '429K', '595K', '416K', '199K', '505K', '730K', '9.7M', '437K', '579K', '832K', '10M', '637K', '270K', '3.4M', '340K', '403K', '10M']
df["views"] = views
```

```python
df.head()
```

```python
def parse_views(value):
    if pd.isna(value):
        return np.nan

    value = str(value).strip().upper()

    if value.endswith("M"):
        return int(float(value.replace("M", "")) * 1_000_000)

    elif value.endswith("K"):
        return int(float(value.replace("K", "")) * 1_000)

    else:
        return int(float(value))
```

```python
df["views_numeric"] = df["views"].apply(parse_views)
```

```python
df.head()
```

```python
df.drop(columns=["views"], inplace=True)
df.rename(columns={"views_numeric": "views"}, inplace=True)
```
